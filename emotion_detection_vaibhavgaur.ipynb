## Real Time Video demo for Face Emotion Recognition

```python
import cv2
from deepface import DeepFace

# Load the face cascade classifier
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Try different camera indexes (0, 1, 2, etc.)
cap = cv2.VideoCapture(0, cv2.CAP_AVFOUNDATION)

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

print("Camera is working.")

while True:
    ret, frame = cap.read()  # Read a frame from the video feed
    if not ret:
        print("Failed to grab frame.")
        break

    # Convert frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Face detection
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Analyze emotion using DeepFace
    try:
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
        emotion = result[0]['dominant_emotion']  # Extract emotion from the list
    except Exception as e:
        print(f"DeepFace Error: {e}")
        emotion = "Unknown"

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the detected emotion on the video frame
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, emotion, (100, 100), font, 3, (0, 0, 255), 2, cv2.LINE_4)

    # Show the output video frame
    cv2.imshow('Demo Video', frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
